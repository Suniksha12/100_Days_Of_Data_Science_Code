- draw a line in a diagram from position (0,0) to position (8,250).

# **Matplotlib Markers**

- Mark each point with a circle.
- Mark each point with a star.

```python
y = np.array([3,8,1,10])
plt.plot(y,marker = 'o')
plt.show()
```

```python
y = np.array([3,8,1,10])
plt.plot(y,marker='*')
plt.show()
```

```python
y = np.array([3,8,1,10])
plt.plot(y,'o:b')
plt.show()
```

```python
y = np.array([3,8,1,10])
plt.plot(y,marker = 'h',color = 'g')
plt.show()
```

```python
y = np.array([3,8,1,10])
plt.plot(y, marker = 'p', color='r')
plt.show()
```

# **Marker Size,Color**

There are two parameters available to colorize the matplotlib markers:

- Matplotlib Marker Edge Color (mec).

- Matplotlib Marker Face Color (mfc).

```python
y = np.array([3,8,1,10])
plt.plot(y,marker = '*',ms = 25)
plt.show()
```

```python
y = np.array([3,8,1,10])
plt.plot(y ,'o:r', ms = 20, mec = 'b')
plt.show()
```

```python
y = np.array([3,8,1,10])
plt.plot(y, 'o:r' , ms = 20 , mec = 'b', mfc = 'b')
plt.show()
```
