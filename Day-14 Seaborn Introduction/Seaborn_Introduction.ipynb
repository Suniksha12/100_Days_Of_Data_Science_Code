# **Python Seaborn**



It a library mostly used for statistical plotting in Python.

```python
#importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/IRIS.csv")
sns.lineplot(x="sepal_length", y="sepal_width", data=df)
```

# **Using Seaborn with Matplotlib**

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=df)

plt.title("Titke using Matplotlib Function")
plt.show()
```

# **Setting the xlim and ylim**

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=df)

plt.xlim(5)

plt.show()
```

# **Customizing Seaborn Plots**
- Changing Figure Aesthetic

- Removal of Spines

- Changing the figure Size

- Scaling the plots

- Setting the Style Temporarily

- Changing Figure Aesthetic

- set_style() method is used to set the aesthetic of the plot.


```python
sns.lineplot(x="sepal_length", y="sepal_width", data=df)

#changing the theme to dark

sns.set_style("dark")
plt.show()
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=df)

#removing the spines
sns.despine()
plt.show()
```

```python
#changing the figure size
plt.figure(figsize = (3,4))
sns.lineplot(x="sepal_length", y="sepal_width", data=df)

plt.show()
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=df)

#setting the scale of the plot
sns.set_context("paper")

plt.show()
```

```python
#setting the style temporarily
def plot():
  sns.lineplot(x="sepal_length", y="sepal_width", data=df)

with sns.axes_style('darkgrid'):

  plt.subplot(211)
  plot()

plt.subplot(212)
plot()
```

# **Color Palette**

- Colormaps are used to visualize plots effectively and easily.

- One might use different sorts of colormaps for different kinds of plots.

- color_palette() method is used to give colors to the plot.

- Another function palplot() is used to deal with the color palettes and plots the color palette as a horizontal array.

```python
palette = sns.color_palette()
sns.palplot(palette)
plt.show()
```

```python
#diverging color palette

palette = sns.color_palette('PiYG',9)
sns.palplot(palette)

plt.show()
```

```python
#sequential color palette

palette = sns.color_palette('Greens',11)
sns.palplot(palette)
plt.show()
```

```python
def plot():
  sns.lineplot(x="sepal_length", y="sepal_width", data=df)
sns.set_palette('vlag')
plt.subplot(211)

plot()

sns.set_palette('Accent')
plt.subplot(212)
plot()

plt.show()
```

# ***Multiple plots with Seaborn***

Using Matplotlib

- Using add_axes() method

- Using subplot() method

- Using subplot2grid() method

Using Seaborn

- Using FacetGrid() method

- Using PairGrid() method

- Using add_axes() method

```python
def graph():
  sns.lineplot(x="sepal_length", y="sepal_width", data=df)

fig = plt.figure(figsize=(5,4))

#creating first axes fro the figure
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])

#plotting the graph
graph()

#creating second axes for the figure
ax2 = fig.add_axes([0.5,0.5,0.3,0.3])

#plotting the graph
graph()

plt.show()
```

```python
#using subplot
def graph():
  sns.lineplot(x="sepal_length", y="sepal_width", data=df)

plt.subplot(121)
graph()

plt.subplot(122)
graph()

plt.show()
```

```python
#using subplot2grid() method
def graph():
  sns.lineplot(x="sepal_length", y="sepal_width", data=df)

#adding the subplots
axes1 = plt.subplot2grid((7,1),(0,0),rowspan = 2,colspan = 1)
graph()

axes2 = plt.subplot2grid ( (7, 1), (2, 0), rowspan = 2, colspan = 1)
graph()

axes3 = plt.subplot2grid ( (7, 1), (4, 0), rowspan = 2, colspan = 1)
graph()
```

```python
#using facetGrid() method
plot = sns.FacetGrid(df , col="species")
plot.map(plt.plot,"sepal_width")

plt.show()
```

```python
#using PairGrid() method

plot = sns.PairGrid(df)
plot.map(plt.plot)

plt.show()
```
